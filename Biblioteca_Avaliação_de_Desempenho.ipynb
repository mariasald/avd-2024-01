#**Desvio Padrão**

Para calcular o desvio padrão é necessário implementarmos algumas funções

##Média Amostral

A média amostral é uma medida estatística que representa o valor médio dos elementos em uma amostra de dados. A média amostral é calculada somando todos os valores da amostra e dividindo pelo número total de elementos na amostra.

```python
def media_amostral(dados):
    return media_aritmetica(dados)
```

##Variância Amostral

A variância é uma medida estatística que descreve o quão dispersos os valores estão em relação à média amostral. Ela mede a média dos quadrados das diferenças entre cada valor individual e a média amostral

```python
def variancia_amostral(dados):
    n = len(dados)
    media = media_aritmetica(dados)
    soma_diferencas_quad = np.sum(np.fromiter(((x - media) * (x - media) for x in dados), dtype=float))
    variancia = soma_diferencas_quad / n
    return round(variancia)
```

##Desvio Padrão

O desvio padrão é uma medida de dispersão que fornece uma ideia de quão espalhados os valores estão em relação à média amostral. Quanto maior o desvio padrão, maior a dispersão dos valores.

```python
def desvio_padrao (dados):
    variancia = variancia_amostral(dados)
    desvio_padrao = math.sqrt(variancia)
    return round(desvio_padrao)
```

#Quartis


Os quartis dividem um conjunto de dados ordenados em quatro partes iguais. O primeiro quartil (Q1) é o valor que divide o conjunto de dados em 25% inferior e 75% superior. O segundo quartil (Q2), que é também a mediana, divide os dados em 50% inferior e 50% superior. O terceiro quartil (Q3) divide os dados em 75% inferior e 25% superior.

image.png

##Implementação da Função

```python
def calcular_quartis(dados):
    dados_ordenados = sorted(dados)
    n = len(dados_ordenados)
    quartil_1 = dados_ordenados[int(n * 0.25)]
    quartil_2 = dados_ordenados[int(n * 0.5)]
    quartil_3 = dados_ordenados[int(n * 0.75)]
    return quartil_1, quartil_2, quartil_3
```

##Aplicação da Função:

```python
# Dados fictícios
dados = [12, 34, 45, 56, 67, 78, 89, 90, 91, 92, 93]

# Calcula os quartis
q1, q2, q3 = calcular_quartis(dados)

# Imprime os quartis
print("Primeiro Quartil (Q1):", q1)
print("Segundo Quartil (Q2, Mediana):", q2)
print("Terceiro Quartil (Q3):", q3)
```

Isso significa que:

Q1 (primeiro quartil) é 45, o que significa que 25% dos dados são menores ou iguais a 45.
Q2 (segundo quartil ou mediana) é 78, o que significa que 50% dos dados são menores ou iguais a 78.
Q3 (terceiro quartil) é 91, o que significa que 75% dos dados são menores ou iguais a 91.

Podemos concluir que isso fornece uma compreensão da distribuição dos dados e de como eles estão divididos em partes iguais em relação aos quartis.

#Tamanho da Amostra

No Tamanho da Amostra ideia básica é que quanto maior o tamanho da amostra, menor será a largura do intervalo de confiança, proporcionando uma estimativa mais precisa do parâmetro populacional.

image.png

##Implementação da Função

```python
def tamanho_de_amostra(desvio_padrao, largura, nivel_de_confianca):
    confianca = round((nivel_de_confianca/100),2)
    alpha = 1 - confianca
    alpha_barra = round(((1+confianca)/2), 3)
    z_score = scipy.stats.norm.ppf(alpha_barra)
    numerador = (((desvio_padrao*z_score)**2)*4)
    tamanho_amostra = (numerador/(largura**2))
    return int(tamanho_amostra)
```

##Aplicação:

```python
import scipy.stats

def tamanho_de_amostra(desvio_padrao, largura, nivel_de_confianca):
    confianca = round((nivel_de_confianca/100), 2)
    alpha = 1 - confianca
    alpha_barra = round(((1+confianca)/2), 3)
    z_score = scipy.stats.norm.ppf(alpha_barra)
    numerador = (((desvio_padrao*z_score)**2)*4)
    tamanho_amostra = (numerador/(largura**2))
    return int(tamanho_amostra)

# Parâmetros
desvio_padrao = 15
largura_intervalo = 6
nivel_de_confianca = 95

# Calcula o tamanho de amostra necessário
tamanho_amostra_necessario = tamanho_de_amostra(desvio_padrao, largura_intervalo, nivel_de_confianca)

# Imprime o resultado
print("Tamanho de Amostra Necessário:", tamanho_amostra_necessario)
```

*   O desvio padrão é 15.
*   A largura do intervalo desejada é 6.
*   O nível de confiança é 95%.


Ao rodar esse código, você obterá o tamanho de amostra necessário para alcançar um intervalo de confiança com a largura desejada.

Isso significa que, para um desvio padrão de 15, uma largura de intervalo de 6 e um nível de confiança de 95%, você precisaria de uma amostra de tamanho aproximado de 96 para alcançar esses critérios.

##Tamanho de uma Amostra - Se a variância não é conhecida

image.png

##Implementação da Função

```python
def tamanho_de_amostra_var_desconhecida(desvio_padrao_amostra, largura_intervalo, nivel_de_confianca):
    confianca = nivel_de_confianca / 100
    alpha = 1 - confianca
    t_score = scipy.stats.t.ppf(1 - alpha/2, df=math.inf)  # Usando infinitos graus de liberdade

    n_numerador = (t_score * (desvio_padrao_amostra / largura_intervalo)) ** 2
    n_denominador = 4 * (desvio_padrao_amostra / largura_intervalo) ** 2

    tamanho_amostra = math.ceil(max(n_numerador, n_denominador))
    return tamanho_amostra
```

Portanto, a determinação cuidadosa do tamanho da amostra é um passo crítico na obtenção de resultados confiáveis e representativos em análises estatísticas.

##Tamanho da Amostra - Se não conhecemos previamente o tamanho do IC

Basicamente, a ideia é que se desejamos uma estimativa com uma certa porcentagem de precisão (r%), podemos calcular o tamanho da amostra necessário para atingir essa precisão em termos do desvio padrão, da média estimada e do valor r%. A fórmula permite determinar o tamanho mínimo da amostra para alcançar o intervalo de confiança desejado.

image.png

##Implementação da Função

```python
def tamanho_de_amostra_por_precisao(desvio_padrao, nivel_de_confianca, precisao_percentual, media_estimada):
    z_score = scipy.stats.norm.ppf(1 - (1 - nivel_de_confianca) / 2)
    tamanho_amostra = math.ceil((100 * desvio_padrao * z_score) / (precisao_percentual * media_estimada))
    return tamanho_amostra
```

# Teste de Hipótese

O teste de hipótese é uma abordagem estatística usada para tomar decisões sobre uma afirmação feita sobre uma população com base em evidências de uma amostra. No caso dessa abordagem, você estaria testando se os dados da amostra fornecem evidências suficientes para rejeitar a hipótese nula, ou seja, para considerar que a média populacional é diferente de u0.

image.png

##Implementação da Função

```python
def calcula_distribuicao_normal_cdf(z):
    t = 1.0 / (1.0 + 0.2316419 * abs(z))
    coeficientes = [0.31938153, -0.356563782, 1.781477937, -1.821255978, 1.330274429]
    cdf = 1.0 - 1.0 / math.sqrt(2*math.pi) * math.exp(-z**2 / 2.0) * (coeficientes[0]*t + coeficientes[1]*t**2 + coeficientes[2]*t**3 + coeficientes[3]*t**4 + coeficientes[4]*t**5)
    return cdf

def teste_hipotese_z(media_amostral, media_nula, desvio_padrao_populacional, tamanho_amostra, nivel_significancia, alternativa='dois-lados'):

    escore_z = (media_amostral - media_nula) / (desvio_padrao_populacional / math.sqrt(tamanho_amostra))

    if alternativa == 'dois-lados':
        valor_p = 2 * (1 - calcula_distribuicao_normal_cdf(abs(escore_z)))
    elif alternativa == 'menor':
        valor_p = calcula_distribuicao_normal_cdf(escore_z)
    elif alternativa == 'maior':
        valor_p = 1 - calcula_distribuicao_normal_cdf(escore_z)
    else:
        raise ValueError("Hipótese alternativa inválida")

    if valor_p < nivel_significancia:
        return "Rejeitar a hipótese nula"
    else:
        return "Não rejeitar a hipótese nula"
```

A função de distribuição acumulada F(x) para a variável aleatória discreta X  é definida como a probabilidade de que X assuma um valor menor ou igual a x

##Distribuição de probabilidade de X, uma V.A. discreta


O valor esperado (ou média) de uma variável aleatória discreta X é uma medida importante que representa o "centro" da distribuição de probabilidade. Ele é calculado ponderando cada valor possível da variável pelo seu respectivo valor de probabilidade.

image.png

##Implementação da Função

```python
def calcular_valor_esperado_VA_discreta(valores, probabilidades):
    if len(valores) != len(probabilidades):
        raise ValueError("Os tamanhos das listas de valores e probabilidades devem ser iguais.")

    valor_esperado = 0
    for i in range(len(valores)):
        valor_esperado += valores[i] * probabilidades[i]

    return valor_esperado
```

Portanto, o valor esperado da variável aleatória  X é 1.625. Isso significa que, em média, esperamos que os valores de X se concentrem em torno de 1.625.

```python
# Valores da variável aleatória
valores = [-1, 1, 2, 3]

# Probabilidades correspondentes
probabilidades = [1/4, 1/8, 1/8, 1/2]

# Calcular o valor esperado
valor_esperado = calcular_valor_esperado_VA_discreta(valores, probabilidades)

print("O valor esperado da variável aleatória é:", valor_esperado)
```

O valor esperado é uma medida de tendência central e descreve onde a distribuição está "centralizada". Neste caso, como o valor esperado é positivo, isso indica que a maioria dos valores da variável aleatória  X tende a ser maior do que 1.625.

image.png

Vamos implementar os cálculos do valor esperado  E[X2]e da variância Var[X] em Python.

```python
# Valores de x e suas probabilidades associadas
x_values = [-1, 1, 2, 3]
p_values = [1/4, 1/8, 1/8, 1/2]

# Cálculo do valor esperado E[X^2]
E_X2 = sum([(x ** 2) * p for x, p in zip(x_values, p_values)])

# Cálculo do valor esperado E[X]
E_X = sum([x * p for x, p in zip(x_values, p_values)])

# Cálculo da variância Var[X]
Var_X = E_X2 - (E_X ** 2)

print("E[X^2] =", format(E_X2, ".3f"))
print("E[X] =", format(E_X, ".3f"))
print("Var[X] =", format(Var_X, ".3f"))
```

Isso significa que o valor esperado de
X^2 é 5.375, o valor esperado de X é 1.625 e a variância de X é 2.734375. A variância representa a dispersão dos valores da variável aleatória em relação à sua média, fornecendo uma medida da "propagação" dos valores em torno do valor médio.

##**Variável Aleatória Bernoulli**


A distribuição de Bernoulli é um caso especial da distribuição binomial, usada para modelar experimentos binários ou de dois resultados possíveis, como sucesso ou fracasso, sim ou não. Ela descreve uma única tentativa independente, onde o resultado pode ser considerado como "sucesso" (1) com probabilidade  p ou "fracasso" (0) com probabilidade 1−p.

image.png

 Como a distribuição Bernoulli é uma distribuição discreta com apenas dois resultados possíveis, suas medidas de centralidade e dispersão são relativamente simples.

##Implementação da Função

```python
import math

def valor_esperado_bernoulli(p):
    return 0 * (1 - p) + 1 * p

def variancia_bernoulli(p):
    return p * (1 - p)

def coeficiente_variacao_bernoulli(p):
    return math.sqrt(variancia_bernoulli(p)) / valor_esperado_bernoulli(p)
```

##**Variável Aleatória Binomial**


A distribuição binomial é frequentemente utilizada para modelar eventos que têm dois resultados possíveis, como sucesso ou fracasso, sim ou não. Ela é caracterizada pelos parâmetros
n (número de tentativas independentes) e

Essas fórmulas nos fornecem medidas importantes para entender a distribuição e a dispersão dos resultados de uma variável aleatória binomial. O valor esperado representa uma média teórica de resultados, a variância mede a dispersão em torno da média, e o coeficiente de variação fornece uma medida relativa da variação em relação à média

##Implementação da Função

```python
def calcular_valor_esperado(n, p):
    valor_esperado = n * p
    return valor_esperado

def calcular_variancia(n, p):
    variancia = n * p * (1 - p)
    return variancia

def calcular_coeficiente_variacao(n, p):
    coeficiente_variacao = math.sqrt(n * p * (1 - p)) / n
    return coeficiente_variacao
```

##Aplicação da Função

```python
# Parâmetros da distribuição binomial
n = 10  # Número de tentativas
p = 0.3  # Probabilidade de sucesso

# Calculando o valor esperado
esperanca = calcular_valor_esperado(n, p)
print("Valor Esperado:", esperanca)

# Calculando a variância
variancia = calcular_variancia(n, p)
print("Variância:", variancia)

# Calculando o coeficiente de variação
coef_variacao = calcular_coeficiente_variacao(n, p)
print("Coeficiente de Variação:", coef_variacao)
```

Valor Esperado: O valor esperado calculado é aproximadamente 3.0. Isso significa que, em média, espera-se obter cerca de 3 sucessos (ou "eventos de interesse") em um total de 10 tentativas, considerando uma probabilidade de sucesso de 0.3 em cada tentativa.

Variância: A variância calculada é aproximadamente 2.1. A variância representa a dispersão dos valores em relação à média. Nesse caso, a variância de 2.1 indica que os resultados podem variar consideravelmente em torno da média de 3.0, com maior dispersão em comparação com uma distribuição mais uniforme.

Coeficiente de Variação: O coeficiente de variação calculado é aproximadamente 0.458. Esse coeficiente é uma medida relativa da dispersão em relação à média. Um coeficiente de variação menor indica menor dispersão relativa à média. Nesse caso, o valor de 0.458 sugere que a variação é moderada em relação à média.

Em resumo, os resultados mostram que, em uma distribuição binomial com =10
n=10 tentativas e =0.3 p=0.3 probabilidade de sucesso, a média é cerca de 3 sucessos, com uma variação moderada ao redor dessa média. Isso implica que os resultados podem ser relativamente dispersos, mas ainda mantêm uma tendência central próxima da média calculada.

##**Variável Aleatória Geométrica**


A distribuição geométrica modela o número de tentativas necessárias até que ocorra o primeiro sucesso em uma sequência de tentativas independentes e idênticas de Bernoulli, onde cada tentativa tem uma probabilidade de sucesso p.

##Implementações das Funções


```python
def valor_esperado_geometrica(p):
    return 1 / p

def variancia_geometrica(p):
    return (1 - p) / (p ** 2)

def coeficiente_variacao_geometrica(p):
    return math.sqrt(variancia_geometrica(p)) / (1 / p)
```

Portanto, à medida que as tentativas de retransmissão aumentam, a probabilidade de sucesso aumenta.

Essa distribuição é aplicada em diversos contextos, como previsões de ocorrências de eventos raros, como falhas em sistemas, chamadas telefônicas em um call center, acidentes de trânsito em uma determinada região e muitos outros cenários em que contamos a ocorrência de eventos de interesse em um intervalo de tempo ou espaço.

##Implementaçã das Funções

```python
def valor_esperado_poisson(lambd):
    return lambd

def variancia_poisson(lambd):
    return lambd

def coeficiente_variacao_poisson(lambd):
    return math.sqrt(lambd) / lambd
```

##Aplicação

```python
# Parâmetro lambda da distribuição de Poisson
lambd = 3.0

# Calculando e imprimindo os resultados
esperanca = valor_esperado_poisson(lambd)
variancia = variancia_poisson(lambd)
coef_variacao = coeficiente_variacao_poisson(lambd)

print(f"Valor Esperado (E[X]): {esperanca}")
print(f"Variância (Var[X]): {variancia}")
print(f"Coeficiente de Variação (CV[X]): {coef_variacao}")
```

O exemplo que você implementou calcula os valores esperados, variância e coeficiente de variação para uma distribuição de Poisson com um parâmetro (λ) igual a 3.0. Vamos analisar os resultados obtidos:

- Valor Esperado (E[X]): O valor esperado representa a média teórica da distribuição de Poisson. Neste caso, com (\λ = 3.0\), o valor esperado é também 3.0. Isso significa que, em média, esperamos que ocorram 3 eventos raros no intervalo de interesse.

- Variância (Var[X]): A variância mede a dispersão dos valores da distribuição. No caso da distribuição de Poisson, a variância é igual ao parâmetro (λ), ou seja, \(Var[X] = 3.0\). Isso indica que a variação dos eventos raros em torno da média é de 3.0.

- Coeficiente de Variação (CV[X]): O coeficiente de variação fornece uma medida da variabilidade relativa em relação ao valor esperado.

Portanto, concluímos que para uma distribuição de Poisson com (\λ = 3.0\), temos uma média de 3.0 eventos raros, uma variação igual a 3.0 e um coeficiente de variação em torno de 0.577, indicando uma moderada dispersão relativa dos valores em relação à média.

```python
def probabilidade_poisson(lambd, k):
    terms = [(math.exp(-lambd) * (lambd ** i) / math.factorial(i)) for i in range(k + 1)]
    prob = sum(terms)
    return prob
```

```python
# Parâmetros do problema
lambd = 2  # Fluxo de chegadas por minuto
k = 3  # Número máximo de jobs desejado

# Calculando e imprimindo a probabilidade
prob_max_tres_jobs = probabilidade_poisson(lambd, k)

print(f"Probabilidade de chegarem no máximo três jobs: {prob_max_tres_jobs:.3f}")
```

O resultado impresso pelo código representa a probabilidade de ocorrerem no máximo três eventos (jobs) no intervalo de tempo, considerando um fluxo médio de chegadas de 2 jobs por minuto. No exemplo, a probabilidade é calculada como aproximadamente 0.857, o que indica que há uma alta probabilidade de ocorrerem 0, 1, 2 ou 3 jobs no intervalo de tempo.
